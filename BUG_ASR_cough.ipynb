```python
from glob import glob
import random
import json
Cello=glob("/content/Medical-voice-recognition_processing_system/cough/Negative/*.wav")
temp_Cello=[]
temp_BCello=[]
for i in Cello:
  json_temp={'audio_filepath':i,
        'duration':2,
        'label':'N',
        'text':'_',
        'offset':0}
  temp_Cello.append(json_temp)
  json_temp={'audio_filepath':'/content/Medical-voice-recognition_processing_system/background/white_noise.wav',
        'duration':2,
        'label':'N',
        'text':'_',
        'offset':0}
  temp_BCello.append(json_temp)
temp_P=[]
temp_BP=[]
Positive=glob("/content/Medical-voice-recognition_processing_system/cough/Positive/*.wav")
for i in Positive*2:
  json_temp={'audio_filepath':i,
        'duration':2,
        'label':'P',
        'text':'_',
        'offset':0}
  temp_P.append(json_temp)
  json_temp={'audio_filepath':'/content/Medical-voice-recognition_processing_system/background/white_noise.wav',
        'duration':2,
        'label':'P',
        'text':'_',
        'offset':0}
  temp_BP.append(json_temp)
# random.shuffle(temp)
train_temp=temp_Cello[:int(len(temp_Cello)*0.8)]+temp_P[:int(len(temp_P)*0.8)]
train_temp_B=temp_BCello[:int(len(temp_BCello)*0.8)]+temp_BP[:int(len(temp_BP)*0.8)]
test_temp=temp_Cello[int(len(temp_Cello)*0.8):int(len(temp_Cello)*0.9)+1]+temp_P[int(len(temp_P)*0.8):int(len(temp_P)*0.9)+1]
test_temp_B=temp_BCello[int(len(temp_BCello)*0.8):int(len(temp_BCello)*0.9)+1]+temp_BP[int(len(temp_BP)*0.8):int(len(temp_BP)*0.9)+1]
self_temp=[{'audio_filepath':'/content/Medical-voice-recognition_processing_system/cough/Negative/544_Negative_male_27_cough.wav',
      'duration':2,'label':'N','text':'_','offset':0}]
self_tempB=[{'audio_filepath':'/content/Medical-voice-recognition_processing_system/background/white_noise.wav',
      'duration':2,'label':'N','text':'_','offset':0}]
with open('/content/train.json', 'w') as f:
  for i in train_temp:
    json.dump(i, f)
    f.write('\n')
with open('/content/test.json', 'w') as f:
  for i in test_temp:
    json.dump(i, f)
    f.write('\n')
with open('/content/train_background.json', 'w') as f:
  for i in train_temp_B:
    json.dump(i, f)
    f.write('\n')
with open('/content/test_background.json', 'w') as f:
  for i in test_temp_B:
    json.dump(i, f)
    f.write('\n')

with open('/content/self_test.json', 'w') as f:
  for i in self_temp:
    json.dump(i, f)
    f.write('\n')
with open('/content/self_background.json', 'w') as f:
  for i in self_tempB:
    json.dump(i, f)
    f.write('\n')
```

```python
config.model.labels=['N','P']
config.model.train_ds.labels=['N','P']
config.model.validation_ds.labels=['N','P']
config.model.test_ds.labels=['N','P']
config.model.train_ds.manifest_filepath = train_dataset
config.model.validation_ds.manifest_filepath = val_dataset
config.model.test_ds.manifest_filepath = test_dataset
```

```python
import torch
import pytorch_lightning as pl
```

```python
accelerator = 'gpu' if torch.cuda.is_available() else 'cpu'
config.trainer.devices = 1
config.trainer.accelerator = accelerator

# Reduces maximum number of epochs to 5 for quick demonstration
config.trainer.max_epochs = 15

# Remove distributed training flags
config.trainer.strategy = 'auto' #
```

```python
config
```

```python
trainer = pl.Trainer(**config.trainer)
```

```python
from nemo.utils.exp_manager import exp_manager
exp_dir = exp_manager(trainer, config.get("exp_manager", None))
```

```python
exp_dir = str(exp_dir)
exp_dir
```

```python
vad_model = nemo_asr.models.EncDecClassificationModel(cfg=config.model, trainer=trainer)
```
